# Intercorr√©lation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1e3
n = np.arange(0,N)

# Signal test : exponentielle d√©croissante d√©cal√©e
k = 200;
x = np.where( n>=k, np.exp(-(n-k)/100), 0)

# Motif : exponentielle d√©croissante
m = np.where( n>=0, np.exp(-n/100), 0)

# Intercorr√©lation entre x et m
Rxm = np.correlate(x, m, mode="full")
nn = np.arange(-(N-1),N)
imax = np.argmax(Rxm)

# Affichage
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.plot(n,x)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title(f'Signal (k={k})')
plt.subplot(1,3,2)
plt.plot(n,m)
plt.xlabel('n')
plt.ylabel('m[n]')
plt.title('Motif')
plt.subplot(1,3,3)
plt.plot(nn,Rxm)
plt.plot([nn[imax], nn[imax]],[0, Rxm[imax]])
plt.xlabel('n')
plt.ylabel('Rxm[n]')
plt.title(f"Intercorr√©lation entre x et m (max en n={int(nn[imax])})")
plt.show()
```

```python
# tp13.csv
y = np.loadtxt("tp13.csv")

# Intercorr√©lation entre y et m
Rym = np.correlate(y, m, mode="full")
imax = np.argmax(Rym)

# Affichage
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.plot(n,y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Signal bruit√©')
plt.subplot(1,3,2)
plt.plot(n,m)
plt.xlabel('n')
plt.ylabel('m[n]')
plt.title('Motif')
plt.subplot(1,3,3)
plt.plot(nn,Rym)
plt.plot([nn[imax], nn[imax]],[0, Rym[imax]])
plt.xlabel('n')
plt.ylabel('Rym[n]')
plt.title(f"Intercorr√©lation entre y et m (max en n={int(nn[imax])})")
plt.show()
```
